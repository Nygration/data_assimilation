# Basic least squares model #

## Setup
Fit the equation 
    $y = ax + b$
given $N$ observation pairs of $x_i$ and $y_i$.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 8)
```

```python
N = 100                  # number of observations

# 'True' coefficients
a_hat = 5.0
b_hat = 3.0

# Noise to add (the standard deviation of the noise)
sigma = 0.3

# Observations
xi = np.random.rand(N)   # random locations between 0 and 1
yi = a_hat * xi + b_hat + sigma*np.random.randn(N)
```

```python
plt.plot(xi, yi, 'ro')

x = np.linspace(0, 1, 50)
plt.plot(x, a_hat*x + b_hat, '--k')
```

## The cost function

The error is defined as $\epsilon_i = y_i - \hat{y}$ where $\hat{y}$ is the 'Truth'. Now, minimize the function $J = \sum_{i=1}^N \epsilon_i^2$. Substitute the model for $\hat{y}$ to get an expression for $J$ with only the data. Define $J$ as a function called 'cost', to be minimized.

```python
def cost(params):
    a, b = params
    return np.sum((yi - a*xi - b)**2)
```

## Graphical solution
Examine the structure of the cost function directly, by looking at a contour plot of $J$ as a function of the parameters.

```python
a = np.linspace(-20, 20, 201)
b = np.linspace(-20, 20, 201)

costij = np.array([cost((ai, bi)) for bi in b for ai in a])
costij.shape = (201, 201)
plt.contourf(a, b, np.sqrt(costij), 50, cmap=plt.cm.Blues_r)
plt.colorbar()
plt.plot(a_hat, b_hat, 'r+')
plt.xlabel('a')
plt.ylabel('b')
```

## Brute force parameter estimation
Now, let's find the minimum of the cost function using an optimization algorithm from scipy.

```python
import scipy.optimize

a, b = scipy.optimize.fmin_cg(cost, (1.0, 1.0))
a, b
```

## Analytical solution

```python
a = np.sum((xi-xi.mean())*(yi-yi.mean()))/np.sum((xi-xi.mean())**2)
b = yi.mean() - a*xi.mean()
a, b
```
