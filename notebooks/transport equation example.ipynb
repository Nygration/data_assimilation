```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0, 1, 50)
dx = x[1] - x[0]
dt = 0.01
c = dx/dt

I = np.sin(20*x)

def step(u):
    '''return next time step'''
    unew = np.zeros_like(u)
    fac = dt*c/dx
    unew[1:] = u[1:] - fac*(u[1:] - u[:-1])
    return unew

u = np.copy(I)

for n in range(20):
    u = step(u)

plt.plot(x, u)
plt.plot(x, I, '--r')
```

```python
u = np.zeros(10)
u[5] = 1.0

unew = step(u)

plt.plot(u)
plt.plot(unew)
```
